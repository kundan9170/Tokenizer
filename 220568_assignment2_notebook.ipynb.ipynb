## 1. Concepts: Byte-Pair Encoding (BPE)

**Byte-Pair Encoding (BPE)** is an iterative string-compression algorithm originally designed for data compression, but widely adopted in NLP for its elegant approach to subword tokenization.

### 1.1. Motivation

Traditional tokenization splits text using white spaces or punctuation, resulting in a high `<unk>` (unknown word) rate for out-of-vocabulary words during inference. BPE addresses this by segmenting text into frequent **subword units**: short fragments (like morphemes, syllables, or common letter combinations). This means that even rare or novel words can be represented as a combination of known subword tokens, vastly reducing reliance on the `<unk>` token.

### 1.2. Core Algorithm: Step-by-Step

1. **Initialization:**  
   - Start with the full corpus, splitting each word into its characters (optionally with a special end-marker, e.g., `</w>` for each word's end).  
   - The initial vocabulary consists of all unique characters present, plus reserved tokens:
     - `<pad>`, `<unk>`, `<s>`, `</s>`.
   
2. **Main Loop (Iterative Merging):**  
   - **Count Frequencies:** For every adjacent pair (bigram) of symbols in the entire corpus, count how often each occurs.
   - **Select Pair:** Identify the most frequent pair.
     - If multiple pairs have the same frequency, select the lexicographically smallest (i.e., the one that would appear first in dictionary order).
   - **Merge:** Replace every instance of that pair in the corpus with a new symbol denoting their concatenation, and add this new symbol to your vocabulary.
   - **Repeat:** Continue until you reach your target vocabulary size, or no pair occurs more than once.

3. **Termination:**  
   - When your vocabulary reaches the designated size (e.g., 5,000 including the four reserved tokens), stop merging.  
   - The resulting vocabulary lists all subword units, from characters up to whole words for the most frequent items.

### 1.3. Practical Details & Implementation Hints

- **Word Boundary:** Use an explicit marker (e.g., `</w>`, or an underline `▁` at start), ensuring merges never cross word boundaries.
- **Corpus Representation:** Each word should be an array of initial characters (with marker if used), allowing easier pair finding and replacements.
- **Case Sensitivity:** Retain the original case—do not lowercase the text.
- **Reserved Tokens:** The first four tokens of your vocabulary **must be**, in order:  
  1. `<pad>`
  2. `<unk>`
  3. `<s>`
  4. `</s>`
  These cannot be split or merged.
- **Byte-level Variant (Required)**: The initial symbols for tokenization should be individual UTF-8 bytes rather than Unicode characters. Each byte corresponds to one 8-bit unit of the UTF-8 encoded text. This ensures that all text—including any Unicode characters—can be handled byte-by-byte, guaranteeing full coverage without unknown tokens.
- **Special character handling**: You may assume that the dataset does not include tabs, newlines, or null bytes inside words; treat all valid UTF-8 bytes (including those representing ASCII and multi-byte Unicode characters) as valid input symbols for the tokenizer.

## 1. Concepts: WordPiece Tokenization

**WordPiece** is a greedy, likelihood-driven subword segmentation algorithm that powers tokenization in models such as BERT and ALBERT.

### 1.1. Motivation

WordPiece was introduced to overcome vocabulary and coverage limitations in word-based models, especially in agglutinative or morphologically rich languages. The goal is to balance **expressiveness** (capturing meaningful word parts) with **efficiency** (restricting tokenizer vocabulary size for tractable model training).

### 1.2. Vocabulary Construction

- **Initialization:** Start with a vocabulary of individual characters and four reserved special tokens: `<pad>`, `<unk>`, `<s>`, `</s>`.
- **Iterative Greedy Merging:** At each step:
  - Evaluate all possible candidate merges (pairs of existing tokens that appear adjacent in the corpus).
  - For every possible new token, *estimate* the improvement in the corpus's total log-likelihood under a unigram language model were that token to be added.
  - **Select** the candidate that offers the *maximum* increase in log-likelihood (ΔL).
  - Add that new token to the vocabulary and update the tokenization of the corpus accordingly.
- **Continuation Prefix:**  
  - Tokens which *begin* a word appear as is.  
  - Tokens which occur *inside* a word are prefixed with `"##"` to indicate they are not word-initial.  
    - Example: `"unbelievable"` may segment as: `"un"`, `"##bel"`, `"##iev"`, `"##able"`.
  - This allows lossless reconstruction (detokenization) of the original text and disambiguation of word boundaries.
- **Determinism:**  
  - When multiple candidates provide equal ΔL, choose the one that is lexicographically smallest.  
  - Your implementation must yield the same result for repeated runs on the same data.
- **End Condition:**  
  - Continue until the vocabulary reaches the target size (including the four special tokens).
  - The BERT-style default is 30,000 tokens, but for this assignment you must stop precisely at 5,000.

### 1.3. Mathematical Objective

At each step, you extend the vocabulary by the candidate subword that **maximizes the corpus likelihood** under a unigram LM. The objective is:

$$
\Delta L = (f_{\text{new}} - \sum_{i=1}^{k} f_i) \cdot \log \frac{f_{\text{new}}}{N}
$$

- f<sub>new</sub> — frequency of the candidate subword (after merge)
- f<sub>i</sub> — frequencies of the merged units replaced by the new piece
- N — total number of token occurrences in the tokenized corpus

This merges only those subword pairs which improve the model's fit to the training corpus.

### 1.4. OOV and Fallback

Any unknown word is decomposed greedily into the longest matching WordPiece tokens from the vocabulary; only segments not covered are mapped to `<unk>`.

### 1.5. Community Impact

WordPiece’s likelihood-maximizing merges produce subword sets tailored to actual corpus statistics, supporting robust open-vocabulary modeling and superior generalization in neural language models. Its introduction enabled significant progress in multilingual NLP.

1. **Formula derivation:** Derive the likelihood gain—explain exactly how the formula $$\Delta L$$ is obtained and why it makes sense in the context of maximizing the corpus likelihood.

Initially, we calculate the unigram probability using the Maximum Likelihood Estimate (MLE):
$$
p_t = \frac{c_t}{N}
$$


Consequently, the log-likelihood for the entire corpus is calculated as:
$$\log L = \sum_t c_t \log \frac{c_t}{N} = \sum_t c_t \log c_t - N \log N $$When we combine existing tokens to form a new token *x* that appears $f_{\text{new}}$ times by consuming $f_{i}$ instances of the original tokens (which had total counts of $C_{i}$), the resulting change in log-likelihood is:$$ \Delta \log L = f_{\text{new}} \log f_{\text{new}} + \sum_i (C_i - f_i) \log (C_i - f_i) - \sum_i C_i \log C_i $$For the specific scenario where the original tokens are entirely consumed by the merge (meaning their original counts $C_{i}$ equal the number of instances merged, $f_{i}$), the equation simplifies to:$$ \Delta \log L = f_{\text{new}} \log \frac{f_{\text{new}}}{N} - \sum_i f_i \log \frac{f_i}{N} $$To improve computational speed, an additional approximation is introduced, leading to the following compact scoring function:$$ \Delta L \approx \left( f_{\text{new}} - \sum_i f_i \right) \cdot \log \frac{f_{\text{new}}}{N}$$

This scoring metric prioritizes merges that result in high-frequency new tokens, as this is the primary way to increase the unigram model's likelihood.

## 1. Concepts: Unigram Language-Model Tokenizer

The **Unigram Tokenizer** is a probabilistic subword tokenizer that seeks the best set of subword units (tokens) by maximizing the likelihood of the training corpus under a unigram language model. Unlike BPE and WordPiece—which grow vocabulary through merging—Unigram starts with a large set of candidate tokens and iteratively prunes it down, focusing on tokens that contribute most to representing the text.

### 1.1 Motivation

- **Flexible Segmentation:**  
  Unigram, by treating each subword as independent, allows for multiple possible segmentations of each word, offering greater flexibility and robustness.

### 1.2 Modeling & Training

**You will use a simplified version as described below:**

- **Initialization:**  
  - Start with an oversized seed vocabulary, ` V_s `, that includes all characters as well as frequently occurring substrings (e.g., all substrings up to length 10 that occur at least twice in the corpus). **Initialize with a vocabulary size of 20,000 tokens.**

- **Segment Likelihood Calculation:**  
  - For each word, compute all possible segmentations using available tokens.  
  - The probability of a word is the product of the probabilities of the tokens in one of its segmentations.

- **Corpus Likelihood:**  
  - Define corpus likelihood as the sum of the log-probabilities of all words, where each word may have multiple possible segmentations.

- **Token Pruning Loop:**  
  - **Repeat until you reach the target vocabulary size (e.g., 5,000):**  
    - For each token ` c_i ` in your vocabulary ` V_s `:
        - Temporarily remove ` c_i `, then recalculate the data likelihood (how well the corpus is represented without ` c_i `).  
        - Compute the drop in likelihood due to removing ` c_i ` (i.e., "likelihood loss").
    - Sort all tokens by their associated loss (least loss = least impact).
    - Permanently remove the bottom fraction (e.g., bottom 20%) of least impactful tokens from the vocabulary.  
    - Make sure to **never remove single-character tokens**, to guarantee the tokenizer can always represent any input without `<unk>`.

- **Probability Normalization:**  
  - After each pruning step, renormalize token probabilities so they sum to 1.
